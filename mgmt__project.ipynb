```python
from google.cloud import storage
import pandas as pd

# Define your storage client
storage_client = storage.Client('fa23-i535-sgandhe-suiciderates')

# Define your bucket name
bucket_name = 'suicide_rates_overview'

# Define your file name
file_name = 'master.csv'

# Create a bucket object
bucket = storage_client.get_bucket(bucket_name)

# Create a blob object
blob = bucket.blob(file_name)

# Download the contents of the blob as a string and then parse it using pandas
blob.download_to_filename('/content/' + file_name)
data = pd.read_csv('/content/' + file_name)
```

Now whether to drop 'HDI for year' column or to replace with mean and median or any data i am finding correlation

```python
correlation = data['HDI for year'].corr(data['suicides_no'])

print("The Correlation between 'HDI for year' and 'suicides_no':", correlation)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='HDI for year', y='suicides_no', data=data)
plt.title("Scatter Plot of HDI for year vs Suicides Number")
plt.xlabel("HDI for Year")
plt.ylabel("Number of Suicides")
plt.show()
```

```python
correlation_matrix = data[['HDI for year', 'suicides_no']].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='magma')
plt.title("Heatmap of Correlation between HDI for Year and Number of Suicides")
plt.show()
```

As the correlation between'HDI for year' and 'suicides_no' is very low. I am dropping 'HDI for year' column.

```python
data.drop(['HDI for year','country-year'], axis=1,inplace=True)
```
